```python
data = {'A':[2,4,5,np.nan,78,np.nan,8],
        'B':[12,14,np.nan,78,np.nan,8,55],
        'C':[np.nan,4,5,78,np.nan,8,63],
        'D':[2,4,5,500,78,11,8]}
data
```

**Median is the value that separates the higher half of a data sample or probability distribution from the lower half.**

```python
df['A'].median()
```

** mean() function can be used to calculate mean/average of a given list of numbers.**

```python
df['A'].mean()
```

**The fillna() method replaces the NULL values with a specified value. The fillna() method returns a new DataFrame object unless the inplace parameter is set to True , in that case the fillna() method does the replacing in the original DataFrame instead.**

```python
df['A'].fillna(df['A'].median())
```

**group of operations**

```python
df = pd.read_csv(r'/content/Used_Bikes (4).csv')
df.head()
```

```python
df['brand']
```

**To group data by columns in Pandas, use the groupby() function with the column(s) of interest as arguments. You can then apply an aggregation function as needed.**

```python
grouped = df.groupby('brand')
grouped
```

```python
grouped[['price']]
```

```python
grouped = df.groupby('owner')
grouped
```

```python
grouped[['price']].mean()
```

```python
grouped=df.groupby('brand')
grouped
```

**The agg() method allows you to apply a function or a list of function names to be executed along one of the axis of the DataFrame, default 0, which is the index (row) axis**

```python
grouped['price'].agg(min_price='min',max_price='max',mean_column='mean')
```
